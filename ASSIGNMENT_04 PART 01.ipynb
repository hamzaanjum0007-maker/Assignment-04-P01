```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
std = pd.read_excel("marks_sheet.xlsx")
```

```python
std
```

```python
std["percentage"] = (std["Total"]*100)/600
```

```python
std.round(2)
```

```python
std.to_excel("student_dataset.xlsx")
```

```python
grade_counts = std["Grade"].value_counts()
plt.figure(figsize=(8,4))
plt.pie(grade_counts,
        labels=grade_counts.index, 
        autopct='%1.1f%%'
       )
plt.title("Student Grades")
plt.show()
```

```python
subjects = ["python", "numpy", "pandas", "streamlit", "ML", "DL", "NLP"]
plt.figure();sb.boxplot(data=std[subjects])
plt.title("marks_distirbution"); plt.xticks(rotation=45)
plt.tight_layout(); plt.show()
```

```python
plt.figure();std["Total"].plot(kind ="hist",bins = 100)
plt.title("Total marks"); plt.xlabel ("total") 
plt.ylabel ("no .of students")
plt.tight_layout()
plt.show()
```

```python
subjects = ["python", "numpy", "pandas", "streamlit", "ML", "DL", "NLP"]
heat = std[subjects].corr()
plt.figure(); sb.heatmap(heat, annot=True)
plt.title("Correlation Heatmap — subjects"); plt.tight_layout(); plt.show()
```

```python
plt.figure()
sb.scatterplot(data = std ,x = "python",y = "ML" ,hue = "Grade", s = 125)
plt.title('python vs ML (colored by grade)')
plt.tight_layout()
plt.show()
```
