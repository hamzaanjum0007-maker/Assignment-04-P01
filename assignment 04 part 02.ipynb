```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
emp = pd.read_csv("employee.csv")
```

```python
emp
```

```python
emp[emp.Department.isna()]
```

```python
emp.loc[5 , "Department"] = "sales"
```

```python
emp.loc[15 , "Department"] = "sales"
```

```python
emp
```

```python
emp[emp.Salary.isna()]
```

```python
mean_salary = emp[emp["Designation"] == "Officer"]["Salary"].mean()
```

```python
emp.loc[(emp["Designation"] == "Officer") & (emp["Salary"].isna()),"Salary"] = mean_salary
```

```python
emp
```

```python
mean_manager = emp[emp["Designation"] == "Manager"]["Salary"].mean()
```

```python
emp.loc[(emp["Designation"] == "Manager") & (emp["Salary"].isna()),"Salary"] = mean_manager
```

```python
emp
```

```python
emp["Tax"] = (emp["Salary"] * 0.10)
```

```python
emp
```

```python
emp["annual_salary_bonus"] = (emp["Salary"] * 0.12)
```

```python
emp
```

```python
emp
```

```python
emp.to_csv("employee_2.csv")
```

```python
plt.figure();emp["Salary"].plot(kind ="hist",bins = 30)
plt.title("salary spread sheet"); plt.xlabel ("SALARY") 
plt.ylabel ("frequency")
plt.tight_layout()
plt.show()
```

```python
bar = emp.groupby("Department")["Salary"].mean()
bar.plot(kind = "bar" , color = "skyblue" , edgecolor = "black")
plt.title("avg salary by department")
plt.xlabel = ("Department")
plt.ylabel = ("avg_Salary")
plt.tight_layout()
plt.show()
```

```python
emp.plot(kind="scatter", x="Age", y="Salary", color="purple")
plt.title("Age vs Salary Correlation")
plt.xlabel=("Age (Years)")
plt.ylabel=("Salary")
plt.tight_layout()
plt.show()
```

```python
designation_counts = emp["Designation"].value_counts()
designation_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Employee Designation count ")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='Salary', y='Tax', data=emp)
plt.title("Tax vs Salary")
plt.xlabel=("Monthly Salary")
plt.ylabel=("Tax Amount")
plt.show()
```
